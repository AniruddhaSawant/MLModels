# Model Building

```python
x = df1.drop('Happiness Score',axis=1)
y = df1['Happiness Score']
```

- Separated Features and Label

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
X_scale = scaler.fit_transform(x) #Scale the data
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_scale,y,test_size = 0.25,random_state = 77)
# initialized the train_test_split
```

```python
reg= LinearRegression()
reg.fit(x_train, y_train) # built the model
```

```python
reg.predict(scaler.transform([[1.32629,1.32261,0.90563,0.63297,0.32957,2.45176]])) #Some manual predictions
```

```python
import pickle
```

```python
file = 'World_Happiness_Report.pickle'
pickle.dump(reg, open(file, 'wb'))
```

- Saved the Model

```python
load_model = pickle.load(open(file , 'rb'))
```

```python
reg.score(x_train, y_train)
```

- Trained accuracy score is 99.02%

```python
reg.score(x_test, y_test)
```

- Test accuracy score is 97.50%

# Regularization 

```python
from sklearn.linear_model import Ridge,Lasso,RidgeCV,LassoCV
```

```python
Lassocv = LassoCV(alphas = None, max_iter = 100, normalize = True)
Lassocv.fit(x_train, y_train)
```

```python
alpha = Lassocv.alpha_
alpha
```

```python
lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test, y_test)
```

- Lasso model accurancy score is 97.50%

```python
rdgcv = RidgeCV(alphas = np.arange(0.001,0.1,0.01),normalize = True)
rdgcv.fit(x_train, y_train)
```

```python
rdgcv.alpha_
```

```python
rd_model = Ridge(alpha=rdgcv.alpha_)
rd_model.fit(x_train, y_train)
```

```python
rd_model.score(x_test, y_test)
```

- Ridge model accurancy score is 97.50%
